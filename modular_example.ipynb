```python
import pandas as pd
from matplotlib import pyplot as plt

def read_temperatures_from_file(num_measurements: int):
    data = pd.read_csv("temperatures.csv", nrows=num_measurements)
    temperatures = data["Air temperature (degC)"]
    return temperatures

def compute_statistics(temperatures: list[float], num_measurements: int) -> float:
    mean = sum(temperatures) / num_measurements
    return mean

def plot_temperatures(temperatures, num_measurements):
    plt.xlabel("Number of measurement")
    plt.ylabel("Air temperatures")
    plt.plot(temperatures, "r-")
    plt.axhline(y=mean, color="b", linestyle="--")
    plt.savefig(f"{num_measurements}.png")
    plt.show()
    plt.clf()

for num_measurements in [25, 100, 500]:
    # read data from file
    temperatures = read_temperatures_from_file(num_measurements)

    # compute statistics
    mean = compute_statistics(temperatures, num_measurements) 

    # plot results
    plot_temperatures(temperatures, num_measurements)
```
